#### 1. Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.
Для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]])

```python
import numpy as np  # Импортируем библиотеку NumPy для работы с массивами

def diagonal_nonzero_product(matrix):
    """
    Вычисляет произведение всех ненулевых элементов диагонали матрицы.
    Если все элементы диагонали равны нулю, возвращает 1.
    """
    diag = np.diag(matrix)  # Извлекаем диагональные элементы матрицы
    nonzero_diag = diag[diag != 0]  # Оставляем только ненулевые элементы диагонали
    if nonzero_diag.size == 0:  # Если на диагонали нет ненулевых элементов
        return 1  # Возвращаем 1
    else:
        return np.prod(nonzero_diag)  # Вычисляем произведение всех ненулевых элементов

# Создаем массив X
X = np.array([
    [1, 0, 1],  # Первая строка
    [2, 0, 2],  # Вторая строка
    [3, 0, 3],  # Третья строка
    [4, 4, 4]   # Четвертая строка (лишняя для диагонали в неквадратной матрице)
])

result = diagonal_nonzero_product(X)  # Вызываем функцию для вычисления произведения ненулевых элементов диагонали
result  # Выводим результат
```

#### 2. Дана матрица X и два вектора одинаковой длины i и j. Построить вектор np.array([X[i[0], j[0]], X[i[1], j[1]], ... , X[i[N-1], j[N-1]]]).

```python
def build_vector_from_matrix(X, i, j):
    """
    Создает вектор из элементов матрицы X на основе индексов i и j.
    
    Параметры:
    X: 2D NumPy массив (матрица)
    i: 1D NumPy массив (индексы строк)
    j: 1D NumPy массив (индексы столбцов)
    
    Возвращает:
    1D NumPy массив, содержащий элементы X[i[k], j[k]] для всех k.
    """
    # Используем список для выбора элементов X по заданным индексам
    return np.array([X[i[k], j[k]] for k in range(len(i))])  # Создаем вектор на основе индексов i и j

# Задаем матрицу X
X = np.array([
    [1, 0, 1],  # Первая строка
    [2, 0, 2],  # Вторая строка
    [3, 0, 3],  # Третья строка
    [4, 4, 4]   # Четвертая строка
])

# Определяем массив индексов строк
i = np.array([0, 1, 2, 3])  # Индексы строк: берем элементы из строк 0, 1, 2 и 3

# Определяем массив индексов столбцов
j = np.array([0, 1, 2, 0])  # Индексы столбцов: берем элементы из столбцов 0, 1, 2 и 0

# Вызываем функцию для построения вектора
result_vector = build_vector_from_matrix(X, i, j)

result_vector  # Выводим результат
```

#### 3. Даны два вектора x и y. Проверить, задают ли они одно и то же мультимножество.
Для x = np.array([1, 2, 2, 4]), y = np.array([4, 2, 1, 2])

```python
def are_multisets_equal(x, y):
    """
    Проверяет, являются ли два массива равными как мультимножества.
    
    Параметры:
    x: 1D NumPy массив
    y: 1D NumPy массив
    
    Возвращает:
    True, если мультимножества равны (содержат одинаковые элементы с одинаковой частотой), иначе False.
    """
    return np.array_equal(np.sort(x), np.sort(y))  
    # Сортируем оба массива и проверяем их на равенство
    # Если отсортированные массивы совпадают, мультимножества равны

# Пример данных
x = np.array([1, 2, 2, 4])  # Первый массив
y = np.array([4, 2, 1, 2])  # Второй массив

# Вызываем функцию и выводим результат
print(are_multisets_equal(x, y))  # Вывод: True
```

#### 4. Найти максимальный элемент в векторе x среди элементов, перед которыми стоит нулевой.
Для x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0])

```python
import numpy as np  # Импортируем библиотеку NumPy для работы с массивами

# Исходный массив
x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0])  

# Находим индексы элементов массива x, где значение равно 0, но исключаем последний элемент
indices = np.where(x[:-1] == 0)[0] + 1  
# `x[:-1]` — все элементы массива, кроме последнего
# `np.where(x[:-1] == 0)[0]` — индексы, где значения равны 0
# `+ 1` — сдвиг индексов вперед, чтобы получить элементы, следующие за нулями

# Находим максимальное значение среди элементов, следующих за найденными индексами
max_value = np.max(x[indices])  
# `x[indices]` — элементы массива x, находящиеся по индексам, следующим за нулями
# `np.max()` — вычисляет максимальное значение среди этих элементов

max_value  # Выводим максимальное значение
```

#### 6. Реализовать кодирование длин серий (Run-length encoding). Дан вектор x. Необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.
Пример: x = np.array([2, 2, 2, 3, 3, 3, 5, 2, 2]).

```python
def run_length_encoding(x):
    """
    Реализует кодирование длин серий (Run-length encoding).
    
    Параметры:
    x: 1D NumPy массив
    
    Возвращает:
    tuple из двух массивов:
    - первый содержит уникальные элементы;
    - второй содержит количество их повторений.
    """
    numbers = []  # Список для хранения уникальных чисел
    counts = []  # Список для хранения количества повторений каждого числа
    
    current_number = x[0]  # Устанавливаем текущее число как первый элемент массива
    current_count = 1  # Инициализируем счетчик для текущего числа
    
    for i in range(1, len(x)):  # Проходим по массиву начиная со второго элемента
        if x[i] == current_number:  # Если текущий элемент равен предыдущему числу
            current_count += 1  # Увеличиваем счетчик повторений
        else:
            numbers.append(current_number)  # Добавляем текущее число в список уникальных чисел
            counts.append(current_count)  # Добавляем количество повторений в список
            current_number = x[i]  # Обновляем текущее число
            current_count = 1  # Сбрасываем счетчик повторений
    
    # Добавляем данные для последней серии
    numbers.append(current_number)  # Добавляем последнее число в список уникальных чисел
    counts.append(current_count)  # Добавляем его количество повторений в список
    
    return np.array(numbers), np.array(counts)  # Возвращаем результаты в виде NumPy массивов

# Пример
x = np.array([2, 2, 2, 3, 3, 3, 5, 2, 2])  # Исходный массив

# Применяем функцию
numbers, counts = run_length_encoding(x)

# Выводим результаты
print(numbers, counts)  # numbers: [2, 3, 5, 2], counts: [3, 3, 1, 2]
```

#### 7. Даны две выборки объектов - X и Y. Вычислить матрицу евклидовых расстояний между объектами. Сравнить с функцией scipy.spatial.distance.cdist.

```python
from scipy.spatial.distance import cdist  # Импортируем функцию cdist для вычисления матриц расстояний

def euclidean_distance_matrix(X, Y):
    """
    Вычисляет матрицу евклидовых расстояний между двумя наборами точек.
    
    Параметры:
    X: 2D массив NumPy (n_samples_x, n_features) — набор точек.
    Y: 2D массив NumPy (n_samples_y, n_features) — другой набор точек.
    
    Возвращает:
    2D массив (n_samples_x, n_samples_y), где элемент [i, j] — евклидово расстояние между X[i] и Y[j].
    """
    return np.sqrt(np.sum((X[:, np.newaxis, :] - Y[np.newaxis, :, :]) ** 2, axis=2))
    # X[:, np.newaxis, :] — добавляем ось, чтобы получить разницу между каждой точкой из X и всеми точками из Y
    # Y[np.newaxis, :, :] — аналогично, добавляем ось для правильного вычитания
    # (X[:, np.newaxis, :] - Y[np.newaxis, :, :]) — разности координат точек
    # ** 2 — возводим разности в квадрат
    # np.sum(..., axis=2) — суммируем квадраты разностей по координатам (для каждой пары точек)
    # np.sqrt(...) — извлекаем квадратный корень, чтобы получить евклидово расстояние

# Генерируем два случайных набора точек
X = np.random.rand(5, 3)  # 5 точек в 3-мерном пространстве
Y = np.random.rand(4, 3)  # 4 точки в 3-мерном пространстве

# Вычисляем матрицу расстояний вручную
distance_matrix_manual = euclidean_distance_matrix(X, Y)

# Вычисляем матрицу расстояний с использованием функции cdist
distance_matrix_cdist = cdist(X, Y, 'euclidean')  # 'euclidean' указывает метрику расстояния

# Сравниваем матрицы, вычисленные вручную и с помощью cdist
print(np.allclose(distance_matrix_manual, distance_matrix_cdist))  
# np.allclose проверяет, равны ли две матрицы с учетом численных погрешностей. Вывод: True или False.
```
