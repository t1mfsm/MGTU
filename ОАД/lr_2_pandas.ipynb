#### 5. Какие  blended-курсы для четверокурсников существуют? На какой blended-курс записалось наибольшее количество студентов? На каком из blended-курсов собрались студенты с самым высоким средним рейтингом (выведите этот курс и количество студентов на нем)? 

```python
# 1. Фильтруем студентов, которые являются четверокурсниками (например, поступившие в 2020 году)
fourth_year_students = df_latest[df_latest['17_group'] != 0.0]
# Это условие отбирает студентов, у которых значение в столбце '17_group' не равно 0, предполагая, что 0 - это нечетверокурсники или студенты без указанной группы.

# 2. Фильтруем blended-курсы среди четверокурсников
blended_courses = fourth_year_students[fourth_year_students['blended'].notna()]
# Отбираются только те студенты, которые записаны на курсы с типом 'blended', где значения в столбце 'blended' не являются пропущенными (NaN).

# 3. Находим курс, на который записалось наибольшее количество студентов
course_counts = blended_courses['blended'].value_counts()  # Подсчитываем количество студентов по каждому курсу
max_course = course_counts.idxmax()  # Находим курс с наибольшим количеством студентов (индекс максимального значения)
max_course_count = course_counts.max()  # Находим максимальное количество студентов на курсе

# 4. Находим курс с самым высоким средним рейтингом студентов
avg_ratings = blended_courses.groupby('blended')['rating'].mean()  # Считаем средний рейтинг студентов для каждого курса
max_avg_rating_course = avg_ratings.idxmax()  # Находим курс с максимальным средним рейтингом (индекс максимального значения)
max_avg_rating = avg_ratings.max()  # Получаем само максимальное значение среднего рейтинга
max_avg_rating_students_count = blended_courses[blended_courses['blended'] == max_avg_rating_course].shape[0]  # Считаем, сколько студентов на этом курсе

# Выводим результаты
print(f"Blended-курсы для четверокурсников: {blended_courses['blended'].unique()}")  # Выводим все уникальные курсы 'blended', на которые записались четверокурсники
print(f"\nКурс с наибольшим количеством студентов: {max_course} ({max_course_count} студентов)")  # Выводим курс с максимальным количеством студентов и его численность
print(f"\nКурс с самым высоким средним рейтингом: {max_avg_rating_course} с рейтингом {max_avg_rating:.2f} ({max_avg_rating_students_count} студентов)")  # Выводим курс с максимальным средним рейтингом и количество студентов
```

#### 6. Выясните, есть ли студенты с абсолютно одинаковыми предпочтениями по всем курсам.

Для этого сформируйте таблицу, где для каждого возможного набора курсов посчитано количество студентов, выбравших такой набор, и оставьте только строки где это количество больше 1.

В данном случае набор курсов задается упорядоченным множеством ('fall_1', 'fall_2', 'fall_3', 'spring_1', 'spring_2', 'spring_3', 'blended'). Элемент blended будет нулевым для 3-го курса.

```python
# 1. Формируем набор курсов для каждого студента
course_columns = ['fall_1', 'fall_2', 'fall_3', 'spring_1', 'spring_2', 'spring_3', 'blended']  # Указываем названия столбцов с курсами
df_latest['course_set'] = df_latest[course_columns].apply(lambda row: tuple(row), axis=1)  # Для каждого студента создаем кортеж из его курсов, объединяя значения по указанным столбцам

# 2. Группируем по набору курсов и считаем количество студентов для каждого набора
course_counts = df_latest['course_set'].value_counts().reset_index()  # Подсчитываем, сколько студентов имеют одинаковый набор курсов
course_counts.columns = ['course_set', 'student_count']  # Переименовываем столбцы для удобства

# 3. Оставляем только те наборы, где количество студентов больше 1
duplicates = course_counts[course_counts['student_count'] > 1]  # Отбираем только те наборы курсов, которые повторяются более чем у одного студента

# Выводим результат
duplicates  # Выводим таблицу с дублирующимися наборами курсов и количеством студентов
```

#### 7. Найдите курсы по выбору, на которые записывались как студенты 18-го года набора, так и студенты 17-го года.

```python
file_path = 'end_seminar.xlsx'  # Указывает путь к файлу Excel, который будет загружен
df = pd.read_excel(file_path)  # Загружает данные из Excel-файла в DataFrame df

# 1. Отбираем студентов 17-го года набора (группы с номерами от 171 до 177)
students_17_group = df[df['17_group'].between(171, 177, inclusive='both')]  # Фильтруем студентов, у которых значение в столбце '17_group' находится в интервале от 171 до 177

# 2. Отбираем студентов 18-го года набора (группы с номерами от 181 до 188)
students_18_group = df[df['18_group'].between(181, 188, inclusive='both')]  # Фильтруем студентов, у которых значение в столбце '18_group' находится в интервале от 181 до 188

# 3. Формируем множество курсов для студентов 17-го года набора
courses_17_set = set(students_17_group[['fall_1', 'fall_2', 'fall_3', 'spring_1', 'spring_2', 'spring_3']].stack())
# Используем метод stack для "разворачивания" данных по столбцам, чтобы получить все курсы в одном столбце.
# После этого преобразуем полученную серию в множество (set), чтобы избавиться от повторений.

# 4. Формируем множество курсов для студентов 18-го года набора
courses_18_set = set(students_18_group[['fall_1', 'fall_2', 'fall_3', 'spring_1', 'spring_2', 'spring_3']].stack())
# То же самое делаем для студентов 18-го года набора.

# 5. Находим общие курсы между студентами 17-го и 18-го года набора
common_courses = courses_17_set.intersection(courses_18_set)
# Используем метод intersection для нахождения пересечения между двумя наборами курсов.

# 6. Удаляем курсы с кодом 0 и None из множества (если такие есть)
common_courses.discard(0)  # Убираем курс с кодом 0, если он есть в множестве
common_courses.discard(None)  # Убираем курс с значением None, если он есть в множестве

# 7. Выводим результаты
print("Курсы по выбору, на которые записывались студенты 17-го и 18-го года набора:")
print(common_courses)
# Выводим список общих курсов, на которые записались студенты 17-го и 18-го года набора.
```

```python
def get_fall_courses_count(row):
    """Определяет количество осенних курсов, которые должен выбрать студент."""
    if row['17_group'] != 0:  # Проверяем, является ли студент частью 17-го года набора
        if row['17_group'] in (171, 172, 173):  # Если группа студента 171, 172 или 173 (специализация MOP или TI)
            return 2  # Эти студенты должны выбрать 2 курса
        else:
            return 1  # Все остальные студенты 17-го года набора выбирают 1 курс
    else:  # Если студент относится к 18-му году набора
        if row['is_mi']:  # Проверяем, относится ли студент к специализации MI
            return 2  # Студенты специализации MI выбирают 2 курса
        else:
            return 1  # Все остальные студенты 18-го года набора выбирают 1 курс

# Применяем функцию к каждой строке DataFrame и создаем новый столбец 'fall_courses_count'
df_latest['fall_courses_count'] = df_latest.apply(get_fall_courses_count, axis=1)
```
